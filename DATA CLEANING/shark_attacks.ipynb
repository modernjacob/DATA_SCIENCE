# Data Cleaning
This project is to showcase some data cleaning techniques from start to finish. This dataset is quite messy and I will attempt to clean it to a point that it could be used for analysis.

```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/jacobhuckleberry/Desktop/Code/Data Science/Shark Attacks/shark_attacks.csv', encoding='latin1')
df.head()
```

## Columns
To start there are many columns which are likely useless for most analysis purposes, so I will drop these.

```python
# check columns
df.columns

# make columns names uppercase
df.columns = df.columns.str.upper()

# drop unnecessary columns
df = df.drop(['CASE NUMBER', 'AGE', 'NAME', 'TIME', 'INVESTIGATOR OR SOURCE', 'PDF', 'HREF FORMULA', 'HREF', 'CASE NUMBER.1', 'CASE NUMBER.2', 'ORIGINAL ORDER'], axis=1)
```

```python
# Rename columns
df.head()
df = df.rename(columns={'FATAL (Y/N)': 'FATAL'})
```

```python
# Check shape
df.shape
```

```python
df.head()
```

## NULL Values

```python
# check for null values
df.isnull().sum()
```

```python
df = df.fillna('')
```

```python
df.head()
```

```python
df.info()
```

## Go over each Column

```python
# FORMAT DATA COLUMN
df.DATE = df.DATE.apply(lambda x: str(x))
```

```python
df.DATE = df.DATE.str.replace('[^0-9]', '', regex=True)
```

```python
df.DATE = df.DATE.apply(lambda x: x[0:4] + '-' + x[4:6] + '-' + x[6:8])
```

```python
df.DATE.unique()
```

```python
# fix YEAR column
df.YEAR = df.YEAR.apply(lambda x: str(x))
df.YEAR = df.YEAR.str.rstrip('.0')
```

```python
df.YEAR.unique()
```

```python
df.shape
df.YEAR = df.YEAR.replace('', '0000')
```

```python
# drop rows in YEAR column than are not 4 characters long
df = df.drop(df[df.YEAR.str.len() < 4].index)
df.YEAR.unique()
```

```python
df.shape
df.YEAR = df.YEAR.replace('0000', '')
```

```python
df.TYPE = df.TYPE.str.lower()
df.TYPE = df.TYPE.replace('boat', 'boating')
df.TYPE.unique()
df.query('TYPE == "invalid"')

# removing invalid types, since some seem to debunk shark involvment
df = df.drop(df[df.TYPE == 'invalid'].index)
```

```python
df.TYPE.unique()
df.query('TYPE == ""')
df.TYPE = df.TYPE.replace('', 'unprovoked')
```

```python
# assign TYPE as category
df.TYPE.unique()
df.TYPE = df.TYPE.astype('category')
```

```python
# Country column
df.COUNTRY.unique()
df.COUNTRY = df.COUNTRY.str.upper()
df.COUNTRY = df.COUNTRY.str.strip()
```

```python
df.COUNTRY = df.COUNTRY.str.replace("/", "|")
df.COUNTRY = df.COUNTRY.str.strip('?')
df.COUNTRY = df.COUNTRY.str.replace('ST. MAARTIN', 'ST. MARTIN')
```

```python
df.AREA = df.AREA.str.upper()
df.AREA = df.AREA.str.strip()
```

```python
# epic
pd.set_option('display.max_colwidth', None)
df.loc[5943]
```

```python
# since there is not too much I can do with the LOCATION column I'll just strip white space and numbers
df.LOCATION.unique()
df.LOCATION = df.LOCATION.str.strip(' 0123456789')
```

```python
# I'll do another strip for white spaces and numbers on the ACTIVITY column
df.ACTIVITY.nunique()
df.ACTIVITY = df.ACTIVITY.str.strip(' 0123456789.')
df.ACTIVITY.unique()
```

```python
# SEX column
# found that the column contained a space error
df = df.rename(columns={"SEX ": "SEX"})

df.SEX.unique()
```

I'm going to assume that the 'N' is meant to be an 'M' and remove both the 'lli' and '.'

```python
df.SEX = df.SEX.str.strip()
df.query('SEX == "lli"')
df.query('SEX == "N"')
df.SEX = df.SEX.replace('lli', '')
df.SEX = df.SEX.replace('N', '')
df.SEX = df.SEX.replace('.', '')
df.SEX.unique()
```

```python
# strip for INJURY column
df.INJURY = df.INJURY.str.strip(' .\1234567890')
```

```python
# FATAL column
df.FATAL.unique()
```

```python
df.FATAL = df.FATAL.str.strip()
df.FATAL = df.FATAL.str.upper()
df.FATAL.unique()
df.query('FATAL == ""')
```

```python
# Change fatal from INJURY column to Y in FATAL column
df.loc[df['INJURY'] == 'FATAL', 'FATAL'] = 'Y'
df.loc[[5805, 5861, 5944]]
```

```python
df.FATAL.unique()
df.FATAL = df.FATAL.replace("", "N")
df.query('FATAL == "2017"')
df.FATAL = df.FATAL.replace("2017", "N")
```

```python
df.FATAL.unique()
df.FATAL = df.FATAL.astype('category')
```

```python
# found that SPECIES has a specing error 
df.SPECIES.unique()
```

```python
# fixed column name, not much I can do here the values are everywhere, this column should probably be dropped for any analysis
df = df.rename(columns={"SPECIES ": "SPECIES"})
df.SPECIES.nunique()
df.SPECIES.unique()
```
